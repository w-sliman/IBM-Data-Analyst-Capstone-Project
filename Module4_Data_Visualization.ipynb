# Hands-on Lab


## Visualizing distribution of data


### Histograms


ConvertedCompPlot a histogram of `ConvertedComp.`


```python
# your code goes here
QUERY = """
SELECT ConvertedComp
FROM master
"""
df= pd.read_sql_query(QUERY,conn)

import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

#set ggplot style
mpl.style.use(['ggplot'])

count, bin_edges= np.histogram(df.dropna())
df.plot(kind='hist', xticks= bin_edges)
plt.show
```

### Box Plots


Plot a box plot of `Age.`


```python
# your code goes here
import seaborn as sns

QUERY = """
SELECT Age
FROM master
"""
df= pd.read_sql_query(QUERY,conn)

sns.boxplot(df)
plt.show()
```

## Visualizing relationships in data


### Scatter Plots


Create a scatter plot of `Age` and `WorkWeekHrs.`


```python
# your code goes here
QUERY= """
SELECT Age, WorkWeekHrs
FROM master
"""
df= pd.read_sql_query(QUERY, conn)

df.plot(kind= 'scatter',x='Age', y= 'WorkWeekHrs', figsize=(10, 6), color='darkblue')
plt.title('Age vs WorkWeekHours')
plt.xlabel('Age')
plt.ylabel('WorkWeekHrs')
plt.show()
```

### Bubble Plots


Create a bubble plot of `WorkWeekHrs` and `CodeRevHrs`, use `Age` column as bubble size.


```python
# your code goes here
QUERY= """
SELECT WorkWeekHrs, CodeRevHrs, Age
FROM master
"""

df= pd.read_sql_query(QUERY, conn)

#normalize age:
df['Age']=(df['Age']-df['Age'].min())/(df['Age'].max()-df['Age'].min())
df.head()
df.plot(kind='scatter',
        x= 'WorkWeekHrs',
        y= 'CodeRevHrs',
        alpha= 0.30,
        color="blue",
        s=df['Age']* 1000,
       )
plt.title('WorkWeekHours vs CodeRevHrs ')
plt.xlabel('WorkWeekHrs')
plt.ylabel('CodeRevHrs')
plt.show()
```

## Visualizing composition of data


### Pie Charts


Create a pie chart of the top 5 databases that respondents wish to learn next year. Label the pie chart with database names. Display percentages of each database on the pie chart.


```python
# your code goes here
QUERY= """
SELECT COUNT(m.Respondent) as numer_of_respondents , ddb.DatabaseDesireNextYear
FROM master as m
LEFT JOIN DatabaseDesireNextYear as ddb
on m.Respondent= ddb.respondent
GROUP BY ddb.DatabaseDesireNextYear
ORDER BY numer_of_respondents DESC
limit 5
"""

df= pd.read_sql_query(QUERY, conn)
df.set_index('DatabaseDesireNextYear',inplace= True)
df.head()
df['numer_of_respondents'].plot(kind='pie',
                                autopct='%1.1f%%',
                                startangle=90, 
                               )

plt.title('Top 5 databases that respondents wish to learn next year')
plt.axis('equal') # Sets the pie chart to look like a circle.
plt.ylabel('Respondents')


plt.show()
```

### Stacked Charts


Create a stacked chart of median `WorkWeekHrs` and `CodeRevHrs` for the age group 30 to 35.


```python
# your code goes here
QUERY= """
SELECT Age,WorkWeekHrs, CodeRevHrs
FROM master
WHERE Age BETWEEN 30 and 35
"""
df= pd.read_sql_query(QUERY,conn)

df= df.groupby('Age').median()

df.plot(kind= 'bar',stacked=True, figsize= (6,4))
plt.title('Median workweek hrs and CodeRevHrs in 30-35 age group')
plt.xlabel('Age')
plt.ylabel('Hours')
plt.ylim(0,55)
plt.show()
```

## Visualizing comparison of data


### Line Chart


Plot the median `ConvertedComp` for all ages from 45 to 60.


```python
# your code goes here
QUERY= """
SELECT Age,ConvertedComp
FROM master
WHERE Age BETWEEN 45 and 60
"""
df2= pd.read_sql_query(QUERY,conn)
df2= df2.groupby('Age').median()
df2.reset_index(inplace= True)
df2.head()


df2.plot(kind='line', x='Age', y='ConvertedComp', figsize=(8,6), legend= None)
plt.title('Median Compensation for all ages from 45 to 60')
plt.xlabel('Age')
plt.ylabel('Compensation')

plt.show()
```

### Bar Chart


Create a horizontal bar chart using column `MainBranch.`


```python
# your code goes here
QUERY= """
SELECT MainBranch, count(MainBranch) as number_of_respondents
FROM master
GROUP BY MainBranch
"""
df_mb= pd.read_sql_query(QUERY, conn)
df_mb.set_index('MainBranch', inplace=True)
df_mb.index=['Software Developer', 'Not a software developer']
df_mb.head()
df_mb.plot(kind='bar', figsize=(6, 4), legend= None)
plt.title('Respondents Main Branch')
plt.ylabel('Number of respondents')
plt.show()
```
